      <h2>CPT Requirements Overview</h2>
     <p>The Create Performance Task (CPT) requires you to build a program and explain how it works.</p>

     <p>Your program MUST include:</p>
      <ul>
      <li><strong>Input:</strong> user interaction (typing, clicking, etc.)</li>
      <li><strong>Output:</strong> result displayed to the user</li>
      <li><strong>List:</strong> used to store multiple related pieces of data</li>
      <li><strong>Procedure:</strong> a function with at least one parameter(variable that receives input)</li>
      <li><strong>Algorithm:</strong> must include sequencing, selection (if), and iteration (loop)</li>
     </ul>

     <p><strong>Key idea:</strong> You are graded on whether these parts are clearly shown and explained.</p>
    </article>

    <article id="quest-skills" class="card">
      <h2>CPT Skills You Must Know</h2>
     <ul>
       <li><strong>Sequencing:</strong> code runs step-by-step</li>
       <li><strong>Selection:</strong> if statements make decisions</li>
       <li><strong>Iteration:</strong> loops repeat through data</li>
       <li><strong>Boolean Expressions:</strong> true/false checks (ex:    <code>typedName == clubNames[i]</code>)</li>
     </ul>

     <p>These skills are required both in your code and in your exam explanations.</p>

<script>
const input = document.getElementById("codeInput");
const output = document.getElementById("output");
const runBtn = document.getElementById("runBtn");
const lang = document.getElementById("lang");

function runCode() {
  const code = input.value;

  if (lang.value === "js") {
    try {
      const result = Function(code)();
      output.srcdoc = `<pre style="padding:10px;">${result ?? "Executed (no output)"}</pre>`;
    } catch (err) {
      output.srcdoc = `<pre style="color:red; padding:10px;">${err}</pre>`;
    }
  }

  else if (lang.value === "html") {
    output.srcdoc = code;
  }

  else {
    output.srcdoc = "<pre>Unsupported language in static mode.</pre>";
  }
}

runBtn.addEventListener("click", runCode);

// Ctrl + Enter to run
input.addEventListener("keydown", (e) => {
  if (e.ctrlKey && e.key === "Enter") {
    runCode();
  }
});
</script>
